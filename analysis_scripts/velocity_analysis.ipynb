```python
scv.settings.verbosity = 3  # show errors(0), warnings(1), info(2), hints(3)
scv.settings.presenter_view = True  # set max width size for presenter view
scv.set_figure_params()
```

```python
### pick sample ###
#G1C4_0LC20 is GC1; G2C11_F717B is GC2; iPS2C1_1DXLC is iPS1; iPS2C4_SYS57 is iPS2
sample='Heart1n2'
file='Heart1n2.loom'
###  clustering that will be used is the one computed through risc###
clusterkey = 'Cluster'
```

```python
loomfile=file
```

```python
adata = scv.read(loomfile, cache=True)
```

```python
adata.var
```

```python
#The samples indexes are as follows:
### fix loom barcode names to match seurat or RISC
import re
bclist = adata.obs.index.tolist()

sampindex = 1

bclistfixed = []
for cell in bclist:
    #remove prefix/suffix
    cellsp = re.sub('possorted_genome_bam_ZNC7J:','Heart1n2_', cell)
    #cellsp = cell.split(sep=':')[1].split('x')[0]
    #add seurat suffix
    cellsp = cellsp +"-1"
    bclistfixed.append(cellsp)
    
    
adata.obs.index = bclistfixed
```

```python
adata.obs['Barcode'] = adata.obs.index
```

```python
adata.obs["Barcode"]
if(len(set(adata.obs["Barcode"])) == len(adata.obs["Barcode"])):

   print("All elements are unique.")
else:
   print("All elements are not unique.")
```

```python
#Lets add the metadata information. In this metadata file I also added the UMAP coordinates obtained in RISC. So i can use it later on for visualization.
integration_md = pd.read_csv('Infart_RISC_metadata.csv')
integration_md
```

```python
#Now filter the df just for this sample. 
is_subset=integration_md['Set']==sample
md_subset = integration_md[is_subset]
md_subset
```

```python
subset_cells = md_subset['Barcode'].astype(str)
```

```python
adata_subset = adata[subset_cells, ]
```

```python
md_subset['Cluster'] = md_subset['Cluster'].astype(str)
```

```python
#Merge metadata to the object
adata_subset.obs = adata_subset.obs.join(md_subset.set_index('Barcode'), on='Barcode').astype('category')
```

```python
#Add RISC UMAP values to multidim obs (obsm). 
# embeddings to multidim obs (obsm)
expression_umap = md_subset[['UMAP1','UMAP2']]
adata_subset.obsm[sample + '_umap'] = expression_umap.values
```

```python
#proportion of splicing in each cluster
scv.pl.proportions(adata_subset, groupby=clusterkey, figsize=(10,8), save=sample + '_' +clusterkey +'_splicing.pdf')
```

```python
#preprocessing
scv.pp.filter_genes(adata_subset, min_shared_counts=20)
scv.pp.normalize_per_cell(adata_subset)
scv.pp.filter_genes_dispersion(adata_subset, n_top_genes=2000)
scv.pp.log1p(adata_subset)
```

```python
#find mean and variance; 
# computed among nearest neighbors in PCA space...
scv.pp.moments(adata_subset, n_pcs=30, n_neighbors=30)
```

```python
#Get the UMAP dimensionality using Louvain statistics
scv.tl.umap(adata_subset, min_dist=0.5, spread=1.0, n_components=2, maxiter=None, alpha=1.0, gamma=1.0, negative_sample_rate=5, init_pos='spectral', random_state=0, a=None, b=None, copy=False, method='umap', neighbors_key=None)
```

```python
#Clustering using velocity data
scv.tl.louvain(adata_subset, resolution=None, random_state=0, restrict_to=None, key_added='louvain', adjacency=None, flavor='igraph', directed=True, use_weights=False, partition_type=None, neighbors_key=None, obsp=None, copy=False)
```

```python
scanpy_umap=adata_subset.obsm['X_umap']
```

Compute velocity

```python
scv.tl.recover_dynamics(adata_subset, n_jobs = 6)
#The computed velocities are stored in adata.layers
```

```python
scv.tl.velocity(adata_subset)
scv.tl.velocity_graph(adata_subset)
```
